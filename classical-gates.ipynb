# Classical gates

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
import matplotlib.pyplot as plt

%matplotlib inline


# Function to create an AND gate circuit
def and_gate_circuit():
    qc = QuantumCircuit(3)
    # Initialize first two qubits as inputs
    qc.x(0)  # Input A = 1
    qc.x(1)  # Input B = 1
    # AND gate using Toffoli (CCX) gate
    qc.ccx(0, 1, 2)
    return qc

# Function to create an OR gate circuit
def or_gate_circuit():
    qc = QuantumCircuit(3)
    # Initialize first two qubits as inputs
    qc.x(0)  # Input A = 1
    qc.x(1)  # Input B = 1
    # OR gate using multi-control NOT gate with auxiliary qubit
    qc.x(0)
    qc.x(1)
    qc.ccx(0, 1, 2)
    qc.x(0)
    qc.x(1)
    return qc

# Function to create an XOR gate circuit
def xor_gate_circuit():
    qc = QuantumCircuit(3)
    # Initialize first two qubits as inputs
    qc.x(0)  # Input A = 1
    qc.x(1)  # Input B = 1
    # XOR gate using CNOT gates
    qc.cx(0, 2)
    qc.cx(1, 2)
    return qc

# Function to visualize the statevector, Bloch sphere, and Q-sphere
def display_quantum_state(qc):
    # Get the statevector from the circuit
    statevector = Statevector.from_instruction(qc)

    # Display the statevector
    print("State Vector:")
    print(statevector)

    # Visualize Bloch Sphere for all qubits
    print("Bloch Sphere Representation:")
    plot_bloch_multivector(statevector)
    plt.show()  # Ensure plot is displayed

    # Visualize the Q-Sphere
    print("Q-Sphere:")
    plot_state_qsphere(statevector)
    plt.show()  

# Main execution
print("AND Gate Implementation")
and_circuit = and_gate_circuit()
print(and_circuit)
display_quantum_state(and_circuit)

print("\nOR Gate Implementation")
or_circuit = or_gate_circuit()
print(or_circuit)
display_quantum_state(or_circuit)

print("\nXOR Gate Implementation")
xor_circuit = xor_gate_circuit()
print(xor_circuit)
display_quantum_state(xor_circuit)
```

# Half Adder

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector
from qiskit.visualization import plot_bloch_multivector, plot_state_qsphere
import matplotlib.pyplot as plt

%matplotlib inline

# Function to create a Half Adder circuit
def half_adder_circuit():
    qc = QuantumCircuit(3)  # 2 input qubits and 1 output qubit for carry

    # Input values: A = 1, B = 1
    qc.x(0)  # A = 1
    qc.x(1)  # B = 1

    # Half Adder Logic
    qc.cx(0, 2)  # XOR operation for Sum
    qc.cx(1, 2)
    qc.ccx(0, 1, 2)  # AND operation for Carry

    return qc

# Function to visualize the statevector, Bloch sphere, and Q-sphere
def display_quantum_state(qc):
    # Get the statevector from the circuit
    statevector = Statevector.from_instruction(qc)

    # Display the quantum circuit
    print("Quantum Circuit:")
    print(qc.draw())

    # Display the statevector
    print("State Vector:")
    print(statevector)

    # Visualize Bloch Sphere
    print("Bloch Sphere Representation:")
    plot_bloch_multivector(statevector)
    plt.show()

    # Visualize Q-Sphere
    print("Q-Sphere:")
    plot_state_qsphere(statevector)
    plt.show()

# Main execution
print("Half Adder Implementation")
half_adder = half_adder_circuit()
display_quantum_state(half_adder)
```

```python
from qiskit import QuantumCircuit
from qiskit.quantum_info import Statevector

# Create a quantum circuit with 3 qubits
qc = QuantumCircuit(3)

# Encode inputs (adjust this to test different inputs)
# Input A = |1>, Input B = |1>
qc.x(0)  # Set qubit 0 (A) to |1>
qc.x(1)  # Set qubit 1 (B) to |1>

# Apply the OR gate logic
qc.ccx(0, 1, 2)  # Toffoli gate: If A and B are |1>, set qubit 2 to |1>
qc.cx(0, 2)      # CNOT gate: If A is |1>, flip qubit 2
qc.cx(1, 2)      # CNOT gate: If B is |1>, flip qubit 2

# Get the statevector of the circuit
state = Statevector.from_instruction(qc)

# Print the results
print("Final quantum state:", state)
qc.draw()
```
